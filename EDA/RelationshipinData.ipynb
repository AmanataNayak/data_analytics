```python
## relationship between age at marriage and education level
divorce = pd.read_csv('divorce.csv')
divorce['marriage_date'] = pd.to_datetime(divorce['marriage_date'])
divorce['dob_man'] = pd.to_datetime(divorce['dob_man'])
divorce['dob_woman'] = pd.to_datetime(divorce['dob_woman'])
divorce['marriage_age_man'] = divorce['marriage_date'].dt.year - divorce['dob_man'].dt.year
divorce['marriage_age_woman'] = divorce['marriage_date'].dt.year - divorce['dob_woman'].dt.year
```

```python
sns.scatterplot(data=divorce,x='marriage_age_woman', y='marriage_age_man')
plt.show()
```

```python
sns.scatterplot(data=divorce, x='marriage_age_woman', y='marriage_age_man', hue='education_man')
```
