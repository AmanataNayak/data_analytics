## Correlation 
#### Corelation coefficient
1. Quantifies the linear relationship between two variables.
2. Number between -1 and 1.
3. Magnitude corresponds to strength of relationship.
4. Sing(+ or -) correspondes to direction of relationship.

-> (+) as x increase y increase
-> (-) as x increase u decrease

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
world_happiness = pd.read_csv('Data/world_happiness.csv')
world_happiness.head()
```

```python
sns.scatterplot(data=world_happiness,x='happiness_score',y='life_exp')
plt.show()
```

```python
## trend line
sns.lmplot(data=world_happiness,x='happiness_score',y='life_exp',ci=None)
plt.show()
```

```python
world_happiness['life_exp'].corr(world_happiness['happiness_score'])
```

## Correlation caveats

#### transfomration
1. Log tranformation(log(x))
2. Square root transformation (sqrt(x))
3. Reciprocal transformation (1/x)

```python
sns.scatterplot(data=world_happiness,x='gdp_per_cap',y='life_exp')
plt.show()
```

```python
world_happiness['gdp_per_cap'].corr(world_happiness['life_exp'])
```

```python
## log per capita
world_happiness['log_gdp_per_cap'] = np.log(world_happiness['gdp_per_cap'])
world_happiness['log_gdp_per_cap']
```

```python
sns.scatterplot(data=world_happiness,x='log_gdp_per_cap',y='happiness_score')
plt.show()
```

```python
world_happiness['log_gdp_per_cap'].corr(world_happiness['happiness_score'])
```

## Design of experiments
1. Experiment to answer what is the effect of the treatment of response.
2. Treatment: explanatory/indepndent variable.
3. Response: response/dependent variable

#### Controlled expermients
1. Particpations are sssigned by researchers to either treatment group or control group.
2. Group shold be comprable so that causation can be inferred.
3. If groups are not comparable, this could lead to confounding (bias).
