```python
biz_owner = pd.read_pickle('../Data/JoiningData/business_owners.p')
biz_owner.head()
```

```python
license_owner = license.merge(biz_owner,on='account',suffixes=('_lic','_biz'))
license_owner
```

```python
counted_df = license_owner.groupby('title').agg({'account':'count'})
counted_df
```

```python
sorted_df = counted_df.sort_values(by='account',ascending=False)
sorted_df.head()
```

## Merging Multiple DataFrame

```python
ridership = pd.read_pickle('../Data/JoiningData/cta_ridership.p')
cal = pd.read_pickle('../Data/JoiningData/cta_calendar.p')
```

```python
ridership_cal = ridership.merge(cal,on=['year','month','day'])
ridership_cal
```

```python
stations = pd.read_pickle('../Data/JoiningData/stations.p')
stations.head()
```

```python
ridership_cal_stations = ridership.merge(cal,on=['year','month','day']).merge(stations,on='station_id')
ridership_cal_stations
```

```python
##filter
filter_criteria = (
    (ridership_cal_stations['month']==7) &
    (ridership_cal_stations['day_type']=='Weekday') &
    (ridership_cal_stations['station_name'] == 'Wilson')
)
```

```python
ridership_cal_stations.loc[filter_criteria,'rides'].sum()
```

```python
zips = pd.read_pickle('../Data/JoiningData/zip_demo.p')
zips.head() 
```

```python
license_zip_ward = license.merge(zips,on='zip').merge(wards,on='ward')
license_zip_ward.head()
```

```python
license_zip_ward.groupby('alderman').agg({'income':'mean'})
```

```python
license_zip_ward.groupby('alderman').mean('income')
```

```python
land_use = pd.read_pickle('../Data/JoiningData/land_use.p')
land_use.head()
```

```python
land_cen_lic = land_use.merge(cenus,on='ward').merge(license,on='ward',suffixes=['_cen','_lic'])
land_cen_lic.head()
```

```python
pop_vac_lic = land_cen_lic.groupby(['ward','pop_2010','vacant']).agg({'account':'count'})
pop_vac_lic
```

```python
sorted_pop_vac_lic = pop_vac_lic.sort_values(['vacant','account','pop_2010'],ascending=[False,True,True])
sorted_pop_vac_lic.head()
```
