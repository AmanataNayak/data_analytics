```python
import pandas as pd
```

## merge_orderd

```python
gdp = pd.read_csv('../Data/JoiningData/WorldBank_GDP.csv')
gdp
```

```python
sp500 = pd.read_csv('../Data/JoiningData/S&P500.csv')
sp500
```

```python
gdp_sp500 = pd.merge_ordered(gdp,sp500,left_on='Year',right_on='Date',how='left')
gdp_sp500
```

```python
gdp_sp500 = pd.merge_ordered(gdp,sp500,left_on='Year',right_on='Date',how='left',fill_method='ffill')
gdp_sp500
```

```python
gdp_sp500[['GDP','Returns']].corr()
```

```python
pop = pd.read_csv('../Data/JoiningData/WorldBank_POP.csv')
pop
```

```python
ctry_date = pd.merge_ordered(gdp,pop,on=['Year','Country Name'],fill_method='ffill')
ctry_date
```

## merge_asof()
1. simmilar to merge_ordered left join
2. Merged on the nearset key columns and not exact matches. Merged on columns must be sorted.

```python
df = pd.merge_asof(gdp,pop,on='Year')
df
```

```python
df['GDP'].diff()
```

## Selecting data with .query()
1. Accepts an input string used to determine which rows are returned
2. input string is simmilar to where clause in SQL statement

```python
df.query('Year >2015')
```

## Reshaping data with .melt()
Use to convert wide format to long format beacuse long format is more accessible for computers to work with

```python
df = {
    'country':['Brazil','Canada','France','India'],
    'indicator':['Inflation %','Inflation %','Inflation %','Inflation %'],
    '2017':[3.45,1.60,1.03,2.49],
    '2018':[3.66,2.27,1.85,4.86],
    '2019':[3.73,1.95,1.11,7.66]
}

inflation = pd.DataFrame(df)
inflation
```

```python
inflation.melt(id_vars=['country','indicator'],value_vars=['2017','2018','2019'],var_name='year',value_name='annual')
```
