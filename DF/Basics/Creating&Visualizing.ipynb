```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_pickle('Data/Data Manipulation/avoplotto.pkl')
df.head()
```

```python
nb_sold_by_size = df.groupby('size')['nb_sold'].sum()
nb_sold_by_size.plot(kind='bar')
plt.show()
```

```python
nb_sold_by_date = df.groupby('date')['nb_sold'].sum()
nb_sold_by_date.plot(kind='line')
plt.show()
```

```python
df.plot(kind='scatter',x='nb_sold',y='avg_price',title='Number of avocados sold vs Average price')
plt.show()
```

```python
df[df['type']=='conventional']['avg_price'].hist(alpha=0.5,bins=20)
df[df['type']=='organic']['avg_price'].hist(alpha=0.5,bins=20)
plt.show()
```
